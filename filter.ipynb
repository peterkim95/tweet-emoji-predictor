```python
emoji_mappings = {
	"eoji2764": 0,
	"eoji1f60d": 1,
	"eoji1f602": 2,
	"eoji1f495": 3,
	"eoji1f525": 4,
	"eoji1f60a": 5,
	"eoji1f60e": 6,
	"eoji2728": 7,
	"eoji1f499": 8,
	"eoji1f618": 9,
	"eoji1f4af": 15,
	"eoji1f384": 17
}


```

```python
source_file = '20_validation'
output_features_name = 'validation_tweets'
output_label_name = 'validation_labels'

with open('extra_data/'+source_file, 'r') as dfile:
    with open('extra_data/'+output_features_name+'.text', 'w') as tweet_file, open('extra_data/'+output_label_name+'.text', 'w') as labels_file:
        for line in dfile:
            split_line = line.split('\t')
            tweet = (split_line[:1])[0]
            emoji_val = split_line[1:]
            emoji_str = emoji_val[0][:-1]
            
            if emoji_str in emoji_mappings:
                print(tweet+" || "+ str(emoji_mappings[emoji_str]))
                tweet_file.write(tweet+"\n")
                labels_file.write(str(emoji_mappings[emoji_str])+"\n")
            else:
                print("not in dict")
```
